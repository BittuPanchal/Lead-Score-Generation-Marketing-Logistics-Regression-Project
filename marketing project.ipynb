# Label Encoding

```python
# label encoding of variables
df.replace(
          {'DemClusterGroup' : {'C':2, 'D':3, 'F':5, 'A':0, 'B':1, 'E':4, 'U':6},
          'DemGender' : {'U':2, 'F':0, 'M':1},
          'DemReg' : {'Midlands':0, 'North':1, 'South East':3, 'South West':4, 'Scottish':2},
          'DemTVReg' : {'Wales & West':10, 'Midlands':4, 'N West':7,
                         'East':2, 'N East':5,'S & S East':8, 'London':3,
                         'S West':9, 'Yorkshire':11, 'Border':0,'C Scotland':1, 'N Scot':6},
          'LoyalClass' : {'Gold':0, 'Silver':2, 'Tin':3, 'Platinum':1}},
          inplace=True)
df1 = df
```

```python
# drop id column
df1 = df1.drop(['ID'], axis=1)
```

```python
# datset after label encoding and drop id column
df1.head()
```

# Logistics Regression Model

```python
# import libraries for logistics regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# independent Variables
X = df1.drop(['TargetBuy'], axis=1)
```

```python
# dependent variable
Y = df1['TargetBuy']
```

```python
# split dataset into training and testing dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# classification model
classifier = LogisticRegression(solver='lbfgs', max_iter=10000)
```

```python
# fit the training dataset
classifier.fit(X_train, Y_train)
```

```python
# accuracy level
classifier.score(X_train, Y_train)
```

# Confusion matrix

```python
#import confusion matrix library and make confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
y_pred = classifier.predict(X_test)
cm = confusion_matrix(Y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.show()
```

# Prediction

```python
# sample prediction 
pre = (13.0,48.0,3,0,3,8,3,0.01,5.0)
predi = np.asarray(pre)
pred = predi.reshape(1,-1)
prediction = classifier.predict(pred)
if prediction == 1:
    print('potential customer')
else:
    print('not potential customer')
```

```python
# prediction probability 
prediction_x = classifier.predict_proba(X_test)
prediction_x
```

```python
# buying and not buying %age of test dataset
b = pd.DataFrame(Y_test)
c = pd.DataFrame(X_test)
d = pd.concat([c,b],axis=1).reset_index(drop=True)
a = pd.DataFrame(prediction_x, columns=['not_buying_%', 'buying_%'])
df_x = pd.concat([d,a], axis =1)
df_x.head()
```

```python
# prediction probability 
prediction_y = classifier.predict_proba(X_train)
prediction_y
```

```python
# buying and not buying %age of each visitors
f = pd.DataFrame(Y_train)
g = pd.DataFrame(X_train)
h = pd.concat([g,f],axis=1).reset_index(drop=True)
e = pd.DataFrame(prediction_y, columns=['not_buying_%', 'buying_%'])
df_y = pd.concat([h,e], axis =1)
df_y.head()
```

```python
df_xy = pd.concat([a,e]).reset_index(drop=True)
df_xy.head()
```

```python
df_x_y = pd.concat([df,df_xy], axis=1)
df_x_y.head()
```
